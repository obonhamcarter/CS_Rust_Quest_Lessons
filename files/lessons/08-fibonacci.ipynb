## Rust Example

```rust
fn fib(n: usize) -> Vec<u64> {
    let mut seq = vec![0, 1];
    while seq.len() < n {
        let len = seq.len();
        seq.push(seq[len - 1] + seq[len - 2]);
    }
    seq
}
fn main() { println!("{:?}", fib(10)); }
```

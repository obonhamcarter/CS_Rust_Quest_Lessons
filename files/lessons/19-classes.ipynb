## Rust Example

```rust
struct Hero { name: String, hp: i32 }
impl Hero {
    fn describe(&self) { println!("{} has {} HP", self.name, self.hp); }
}
fn main() {
    let hero = Hero { name: String::from("Luna"), hp: 60 };
    hero.describe();
}
```
